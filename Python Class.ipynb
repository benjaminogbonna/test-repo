Python provides many built-in string methods:


````python
s = "hello world"

print(s.upper())    # HELLO WORLD (Converts to uppercase)

print(s.lower())    # hello world (Converts to lowercase)

print(s.title())    # Hello World (Capitalizes each word)

print(s.capitalize()) # Hello world (Capitalizes first letter only)

print(s.replace("hello", "hi"))  # hi world (Replaces a word)

print(s.split())    # ['hello', 'world'] (Splits into a list)

print(s.strip())    # Removes spaces from start and end

print(s.lstrip())    # Removes spaces from left

print(s.rstrip())    # Removes spaces from right

print(s.find("world"))  # 6 (Finds index of substring)

print(s.count("o"))  # 2 (Counts occurrences of 'o')

print(s.islower())  # Checks if the string is in lower case

print(s.isupper())  #  Checks if the string is in upper case

print(s.startswith("hello"))  # Checks if the string starts with the word hello

print(s.endswith("com"))  # Checks if the string ends with the word java

print(s.endswith("world"))  # Checks if the string ends with the word world
````

# Python Lists
A list is an ordered, mutable collection that can store multiple items of different data types. Lists are defined using square brackets "[]" and allow duplicate values. 
It can store different data types, including integers, strings, floats, and even other lists.

### Characteristics of Lists
- Ordered → The elements maintain their order.
- Mutable → Lists can be modified (add, remove, or change elements).
- Heterogeneous → A list can store different data types.
- Indexed → Elements are accessed using an index (starting from 0).

# Tuple

**A tuple is an ordered, immutable collection of elements in Python. Tuples are similar to lists, but they cannot be modified once created. They are defined using parentheses ()**

**Characteristics of Tuples**
- Ordered → The elements maintain their order.
- Immutable → Once created, a tuple cannot be changed (no addition, removal, or modification of elements).
- Heterogeneous → A tuple can store different data types.
- Indexed → Elements are accessed using an index (starting from 0).

### Creating a Tuple
Tuples are created using parentheses (), or simply by separating values with commas.

````python
# Creating different types of tuples
empty_tuple = ()  # Empty tuple
single_element_tuple = (5,)  # Single-element tuple (note the comma)
nums = 5,6,5
numbers = (1, 2, 3, 4, 5)  # Tuple of integers
fruits = ("apple", "banana", "cherry")  # Tuple of strings
mixed = (1, "hello", 3.5, True)  # Tuple with different data types
nested_tuple = ((1, 2, 3), (4, 5, 6))  # Tuple containing other tuples

print(numbers)
````

### Accessing Tuple Elements
You can access tuple elements using indexing and slicing.

````python
fruits = ('apple', 'banana', 'cherry', 'pineapple', 'orange')


print(fruits[0])  
print(fruits[-1])  

# Slicing a tuple
print(fruits[0:2])  
````

### Tuple Immutability
Tuples cannot be modified after creation. If you try to change an element, Python will raise an error.

````python
names = ('Jane', 'Taofeeka', 'Rosheedat', 'Abdulsamad')
numbers[3] = 'Nimah'
````

````python
# If you need a modified version, you can convert the tuple to a list, modify it, and convert it back to a tuple.

names = ('Jane', 'Taofeeka', 'Rosheedat', 'Abdulsamad')
names_list = list(names)  # Convert to list
names_list[3] = 'Nimah'  # Modify the list
names = tuple(names_list)  # Convert back to tuple
print(names) 
````

## When to Use Tuples Instead of Lists?
    Feature	           Tuple	                    List

    Mutability	       Immutable (unchangeable)	    Mutable (changeable)
    Performance	       Faster	                    Slightly slower
    Memory Usage	   Uses less memory	            Uses more memory
    Iteration Speed	   Faster	                    Slightly slower
    Data Security	   More secure	                Less secure


### Use tuples when:
- You want to protect data from modification (e.g., database records, coordinates).
- You need a faster and memory-efficient data structure.

# Set
**A set is an unordered, mutable, and unindexed collection of unique elements in Python. It is used to store multiple values without duplicates.**

**Key Characteristics of Sets**
- Unordered → The elements do not have a specific order.
- Unique → A set cannot contain duplicate values.
- Mutable → You can add or remove elements.
- Unindexed → Elements cannot be accessed using an index like lists or tuples.
- Faster Operations → Due to its implementation using a hash table, set operations like searching are faster.

### Creating a Set
A set is created using curly braces {} or the set() constructor.

````python
empty_set = set()  # Empty set (note: {} creates an empty dictionary, not a set)
fruits = {"apple", "banana", "cherry"}
numbers = {7, 3, 9, 1, 2, 3, 4, 5}
mixed = {1, "hello", 3.5, True}  # A set with mixed data types

print(numbers)
````

````python
# 0 == False
# 1 == True
x = {1, 2, 3, True}
print(x)
````

## Introduction to Python Casting (Type Conversion)

**In Python, casting (also known as type conversion) is the process of converting a variable from one data type to another. Python supports both implicit casting (automatic type conversion) and explicit casting (manual type conversion).**

### 1. Implicit Type Casting (Automatic Conversion)

**Python automatically converts one data type into another when no data loss occurs. This happens mainly with numerical types (integers, floats, etc.).**

**Example of Implicit Casting:**

````python
# Integer is automatically converted to float
num_int = 5
num_float = num_int + 2.5  # Python converts num_int (5) to float (5.0) automatically

print(num_float)  
print(type(num_float))  
````

````python
# Integer is automatically converted to complex
num_complex = num_int + 3j  # Python converts num_int (5) to complex (5+0j)

print(num_complex)  
print(type(num_complex))
````

### Key Takeaways for Implicit Casting:
- Python converts a smaller data type into a larger one automatically.
- Examples include converting int → float and int → complex.
- It prevents data loss and occurs when mathematical operations require a larger data type.

## 2. Explicit Type Casting (Manual Conversion)
Explicit casting is when we manually convert a variable’s data type using built-in functions like:
- int(): Converts a value to an integer.
- float(): Converts a value to a floating-point number.
- str(): Converts a value to a string.
- bool(): Converts a value to a boolean.
- list(), tuple(), set(), dict(), etc.: Converts between different collection types.


### Examples of Explicit Casting

````python
# Converting float to integer (Explicit Casting)
num_float = 7.0
num_int = int(num_float)  # Manually converts float to int (truncates decimal)

print(num_int)  # Output: 7
print(type(num_int))  # Output: <class 'int'>
````

````python
# Converting string to integer
num_str = "123"
num_int = int(num_str)  # Converts string to integer

print(num_int)  # Output: 123
print(type(num_int))  # Output: <class 'int'>
````

````python
# Converting integer to string
age = 25
age_str = str(age)  # Converts integer to string

print(age_str)  # Output: "25"
print(type(age_str))  # Output: <class 'str'>
````

````python
# Converting list to tuple
my_list = [1, 2, 3, 4]
my_tuple = tuple(my_list)  # Converts list to tuple

print(my_tuple)  # Output: (1, 2, 3, 4)
print(type(my_tuple))  # Output: <class 'tuple'>
````

### Key Takeaways for Explicit Casting:
- Used when automatic conversion is not possible.
- Must be done manually using functions like int(), float(), str(), etc.
- Some conversions may lose data (e.g., float to int truncates the decimal part).

## 3. Special Cases in Casting
### String to Integer or Float

Strings can be converted to numbers if they contain valid numerical values.

````python
num_str = "45.6"
num_float = float(num_str)  

print(num_float)  
print(type(num_float))  
````

````python
invalid_str = "45a"
invalid_str = int(invalid_str)  
````

### Boolean Conversion
Python considers:

- 0, "", None, False, and empty collections ([], {}, set()) as False.
- Any non-zero number or non-empty object as True.

````python
print(bool(0))       
print(bool(1))       
print(bool(""))      
print(bool("hello")) 
print(bool([]))      
print(bool([1, 2]))  
````

## Conclusion
**Implicit casting happens automatically when Python converts smaller data types into larger ones (e.g., int → float).** 

**Explicit casting requires manual conversion using functions like int(), float(), str(), bool(), etc.**

**Some conversions may lose information, so be cautious.**

- Comparison (Relational) Operators
- Logical Operators
- Bitwise Operators
- Assignment Operators
- Identity Operators
- Membership Operators

## 1. Arithmetic Operators
These operators are used for basic mathematical operations like addition, subtraction, multiplication, etc.

    Operator	         Description	       Example
    +	                 Addition	           5 + 3 → 8
    -	                 Subtraction	       10 - 4 → 6
    *	                 Multiplication	       7 * 2 → 14
    /	                 Division	           10 / 2 → 5.0 (returns float)
    //	                 Floor Division	       10 // 3 → 3 (ignores decimal part)
    %	                 Modulus (remainder)    10 % 3 → 1
    **	                 Exponentiation (power) 2 ** 3 → 8 (2³)

````python
a = 10
b = 3

print(a + b)   
print(a - b)   
print(a * b)   
print(a / b)   
print(a // b)  
print(a % b)   
print(a ** b)  

print(f'10 divide 3 is {a // b}, and the remainder is {(a % b)} ')
````

## 2. Comparison (Relational) Operators
These operators compare values and return True or False.

    Operator	Description	Example
    ==	Equal to	5 == 5 → True
    !=	Not equal to	5 != 3 → True
    >	Greater than	10 > 3 → True
    <	Less than	2 < 5 → True
    >=	Greater than or equal to	5 >= 5 → True
    <=	Less than or equal to	4 <= 3 → False

````python
x = 10
y = 5

print(x == y)   
print(x != y)   
print(x > y)    
print(x < y)    
print(x >= y)   
print(x <= y)   
````

## 3. Logical Operators
Logical operators are used to combine conditional statements.

    Operator	Description	Example
    and	Returns True if both conditions are true	(x > 5 and y < 10)
    or	Returns True if at least one condition is true	(x > 5 or y > 10)
    not	Reverses the logical state	not(x > 5)

````python
x = 5
y = 10
z = 15
print(x < y or x > z)

print(x > 5 and y < 10) 
print(x > 5 or y > 10)   
print(not(x > 5))        
````

## 4. Bitwise Operators
Bitwise operators perform operations on binary numbers.

    Operator	Description	Example
    &	Bitwise AND	5 & 3 → 1
    `	`	Bitwise OR
    ^	Bitwise XOR	5 ^ 3 → 6
    ~	Bitwise NOT	~5 → -6
    <<	Left shift	5 << 1 → 10
    >>	Right shift	5 >> 1 → 2

````python
a = 5  # Binary: 101
b = 3  # Binary: 011

print(a & b)   # 1 (Binary: 001)
print(a | b)   # 7 (Binary: 111)
print(a ^ b)   # 6 (Binary: 110)
print(~a)      # -6 (Binary: -(101+1))
print(a << 1)  # 10 (Binary: 1010)
print(a >> 1)  # 2 (Binary: 10)
````

## 5. Assignment Operators
Used to assign values to variables.

    Operator	Example	Equivalent To
    =	x = 5	x = 5
    +=	x += 3	x = x + 3
    -=	x -= 2	x = x - 2
    *=	x *= 4	x = x * 4
    /=	x /= 2	x = x / 2
    //=	x //= 3	x = x // 3
    %=	x %= 2	x = x % 2
    **=	x **= 3	x = x ** 3

````python
x = 10
x += 5

print(x)  

x *= 2
print(x)  

x -= 4
print(x)  
````

## 6. Identity Operators
These operators compare memory locations of two objects.

    Operator	Description	Example
    is	Returns True if both variables refer to the same object	x is y
    is not	Returns True if they refer to different objects	x is not y

````python
x = [1, 2, 3]
y = x
z = [1, 2, 3]

print(x is z)     
print(x is not z) 
print(x == z)
````

## 7. Membership Operators
These operators check if a value exists in a sequence (list, tuple, string, etc.).

    Operator	Description	Example
    in	Returns True if a value is in a sequence	"a" in "apple"
    not in	Returns True if a value is not in a sequence	"x" not in "apple"

````python
fruits = ["apple", "banana", "cherry"]

print("apple" in fruits)     
print("mango" not in fruits)
````

## Conclusion
**Python operators allow you to perform various operations on variables and values. Here’s a summary:**
- Arithmetic Operators (+, -, *, /, //, %, **) for mathematical operations.
- Comparison Operators (==, !=, >, <, >=, <=) for value comparisons.
- Logical Operators (and, or, not) for boolean logic.
- Bitwise Operators (&, |, ^, ~, <<, >>) for binary calculations.
- Assignment Operators (=, +=, -=, etc.) for value assignment.
- Identity Operators (is, is not) to compare objects.
- Membership Operators (in, not in) to check for existence in sequences.

##  7. Decorators in Python
**A decorator is a function that modifies or enhances another function without changing its actual code. Decorators "wrap" another function to add extra behavior before or after the original function runs.**


### When and Why Use a Decorator?
- To add common functionality like logging, authentication, timing, etc., without repeating code.
- To modify existing functions in a clean and readable way.
- To follow the DRY principle (Don't Repeat Yourself).


In the example below, the @decorator modifies say_hello().<br/>

wrapper() adds extra behavior before & after execution.

````python
def decorator(func):
    def wrapper():
        func()
        print("After function call")
    return wrapper

@decorator
def say_hello():
    print("Hello!")

say_hello()
````

````python
def decorator(func):
    def cube(x):        
        func(x)
        print(f'After function call, {x} cubed is {x**3}')
    return cube

@decorator
def squared(x):
    nums = [x for x in range(x) if x%2 ==0]
    print(x*x, nums)

squared(15)
````

### Difference Between Static Variable and Instance Variable in Python
**In Object-Oriented Programming (OOP) with Python, variables inside a class can either be instance variables or static (class) variables.**

Use instance variables when each object needs to maintain its own state/data. They are usually defined inside the constructor (__init__) using self.

2. Static (Class) Variables: Static or class variables are shared by all instances of a class. They are defined outside the constructor, usually directly inside the class body. Use static variables when you want to store a value that is common to all instances, like a counter, constant, or shared config. Defined directly in the class, not inside __init__.

````python
class ShoppingCart:
    def __init__(self):
        self.items = []

    def add_item(self, name, price):
        self.items.append({"name": name, "price": price})
        print(f"{name} added to cart.")

    def remove_item(self, name):
        for item in self.items:
            if item["name"] == name:
                self.items.remove(item)
                # print(f"{name} removed from cart.")
                return f"{name} removed from cart."
        print("Item not found in cart.")

    def total_price(self):
        return sum(item["price"] for item in self.items)

    def show_cart(self):
        if not self.items:
            print("Your cart is empty.")
        else:
            print("Items in cart:")
            for item in self.items:
                print(f"{item['name']}: N{item['price']}")
            print(f"Total: N{self.total_price()}")
````

````python
class BankAccount:
    def __init__(self, fullname, acc_type, balance=0.0):
        self.fullname = fullname
        self.acc_type = acc_type
        self.balance = balance

    def deposit(self, amount):            
        if amount > 0:
            self.balance += amount
            print(f"Deposited N{amount:.2f}. New balance: N{self.balance:.2f}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if amount <= 0:
            print("Withdrawal amount must be positive.")
        elif amount > self.balance:
            print("Insufficient funds.")
        else:
            self.balance -= amount
            print(f"Withdrew N{amount:.2f}. New balance: N{self.balance:.2f}")
            
    def transfer(self, account_number, amount):
        if len(account_number)!=10:
            print('Incorrect account number')
        else:
            if amount <= 0:
                print("Transfer amount must be positive.")
            elif amount > self.balance:
                print("Insufficient funds.")
            else:
                self.balance -= amount
                print(f"Transferred N{amount:.2f}. New balance: N{self.balance:.2f}") 
            

    def check_balance(self):
        print(f"Current balance: N{self.balance:.2f}")

    def account_summary(self):
        print(f"Account Holder: {self.fullname}")
        print(f"Account Type: {self.acc_type}")
        print(f"Balance: N{self.balance:.2f}")
````

# Abstraction

**Abstraction in Object-Oriented Programming (OOP) is the concept of hiding internal details and only showing essential features of an object. It helps in reducing complexity and increasing code reusability and flexibility.**

**Think of it as interacting with an interface rather than the implementation.**

### Why and When to Use Abstraction?
**Use it when:**
- You want to define a common interface for different implementations.
- You want to enforce rules that all subclasses must follow.
- You want to separate the "what" from the "how".


**Benefits:**
- Simplifies code structure.
- Increases flexibility and scalability.
- Encourages code consistency across subclasses.


### Abstract Class in Python
**An abstract class:**
- Cannot be instantiated directly.
- Is designed to be inherited.
- May contain one or more abstract methods that must be implemented by subclasses.


### How to Create an Abstract Class
**In Python, abstraction is achieved by:**
- Importing from abc module (Abstract Base Classes)
- Using the ABC class and @abstractmethod decorator

### Example 1

````python
from abc import ABC, abstractmethod

class Shape(ABC):

    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass
````

````python
shape = Shape()
````

````python
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)
````

````python
rect = Rectangle(10, 5)

print("Area:", rect.area())   
````

````python
print("Perimeter:", rect.perimeter())  
````

# Polymorphism

**Polymorphism means "many forms". In Object-Oriented Programming, it allows objects of different classes to be treated as if they are objects of the same class — as long as they share common methods.**

**With polymorphism, the same interface (method or operation) behaves differently depending on the object that uses it.**


### Why Use Polymorphism?
- Promotes code reusability and flexibility.
- Supports the Open/Closed Principle — open for extension, closed for modification.
- Helps write cleaner, extensible, and scalable code.
- Enables dynamic method binding at runtime.


### Types of Polymorphism in Python
- Method Overriding:	Subclass changes the behavior of a method defined in the superclass
- Method Overloading:	One method with different parameters (not natively supported like in Java)
- Constructor Overloading:	Multiple constructors with different parameters (handled manually in Python)
- Operator Overloading:	Using magic methods (dunder methods) to override built-in operators


### 1. Method Overriding
**Method overriding occurs when a child (subclass) defines a method with the same name and signature as a method in its parent (superclass), thereby replacing or extending its behavior.**
**Occurs when a child class provides a different implementation of a method that is already defined in its parent class.**

**When to Use:**<br/>
- When you want to customize or extend the behavior of a method inherited from a parent class.

````python
class Animal:
    def sound(self):
        return "Some sound"

class Dog(Animal):
    def sound(self):
        return "Bark"

class Cat(Animal):
    def sound(self):
        return "Meow"
````

````python
dog = Dog()

dog.sound()
````

````python
cat = Cat()

cat.sound()
````

**PS: Each class overrides the sound() method, but you call the same method on different objects.**


### 2. Method Overloading
**Method overloading means defining multiple methods with the same name but different parameters.**

**Python does not support traditional method overloading like Java or C++.. But you can simulate it with default arguments or *args and **kwargs.**

````python
class Calculator:
    def add(self, a, b=0, c=0):
        return a + b + c
````

### 3. Constructor Overloading (Manual Handling)
**Constructor overloading means having multiple constructors with different sets of parameters.**

**Like methods, Python doesn’t support multiple __init__ methods, but you can simulate it using default parameters or *args.**

````python
class Person:
    def __init__(self, name=None, age=None):
        if name and age:
            print(f"Name: {name}, Age: {age}")
        elif name:
            print(f"Name: {name}")
        else:
            print("No information")
````

### 4. Operator Overloading (Dunder or Magic Methods)
**Python allows you to override built-in operators using special methods like __add__, __str__, __lt__, etc.**

````python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):  # Overload + operator
        return Point(self.x + other.x, self.y + other.y)

    def __str__(self):  # For print()
        return f"({self.x}, {self.y})"
````

````python
p1 = Point(1, 2)
p2 = Point(3, 4)

print(p1 + p2)
````

````python
# An expense Tracker

class Expense:
    def __init__(self, name, amount, category):
        """
        Initialize an expense object
        :param name (str): the name of the expense
        :param amount (float): the amount of the expense
        :param category (str): the category of the expense
        """
        self.name = name
        self.amount = amount
        self.category = category

    def __str__(self):
        """
        String representation on an expense object
        :return (str): formatted string with expense details
        """
        return f'{self.name}: N{self.amount} {self.category}'


class ExpenseTracker:
    def __init__(self, filename='expenses.txt'):
        """
        Initializes an ExpenseTracker object
        :param filename (str):  the file to store expenses
        """
        self.filename = filename
        self.expenses = self.load_expenses()

    def add_expense(self, expense):
        """
        Add an expense to the tracker
        :param expense (Expense): the expense to add
        """
        self.expenses.append(expense)
        self.save_expenses()

    def load_expenses(self):
        """
        Load expenses from the file
        :return list: List of Expense objects
        """
        expenses = []
        try:
            with open(self.filename, 'r') as file:
                for line in file:
                    name, amount, category = line.strip().split(',')
                    expense = Expense(name, float(amount), category)
                    expenses.append(expense)
        except FileNotFoundError:
            pass
        return expenses

    def save_expenses(self):
        """
        Save expenses to the file
        """
        with open(self.filename, 'w') as file:
            for expense in self.expenses:
                file.write(f'{expense.name}, {expense.amount}, {expense.category}\n')

    def display_expenses(self):
        """
        Display all expenses
        """
        for expense in self.expenses:
            print(expense)

    def get_total_expense(self):
        """
        Calculate the total expense amount
        :return (float): total expense amount
        """
        return sum(expense.amount for expense in self.expenses)
````

````python
import tkinter as tk

# Create the main window
root = tk.Tk()
root.title('Simple Calculator')

# Entry wi widget to display the result
entry = tk.Entry(root, width=16, font=('Arial', 24), borderwidth=2, relief='solid')
entry.grid(row=0, column=0, columnspan=4)

# Global variables to store the current input and result
current_input = ''

def button_click(value):
    global current_input
    current_input += str(value)
    entry.delete(0, tk.END)
    entry.insert(0, current_input)

def clear():
    global  current_input
    current_input = ''
    entry.delete(0, tk.END)

def calculate():
    global current_input
    try:
        result = str(eval(current_input))
        entry.delete(0, tk.END)
        entry.insert(0, result)
        current_input = result
    except:
        entry.delete(0, tk.END)
        entry.insert(0, 'Error')
        current_input = ''


buttons = [
    '7', '8', '9', '/',
    '4', '5', '6', '*',
    '1', '2', '3', '-',
    'C', '0', '=', '+'
]

row = 1
col = 0
for button in buttons:
    if button == 'C':
        tk.Button(root, text=button, width=5, height=2, 
                  command=clear).grid(row=row, column=col)
    elif button == '=':
        tk.Button(root, text=button, width=5, height=2, 
                  command=calculate).grid(row=row, column=col)
    else:
        tk.Button(root, text=button, width=5, height=2,
                  command=lambda x=button: button_click(x)
                 ).grid(row=row, column=col)

    col += 1
    if col > 3:
        col = 0
        row += 1

# run the main loop
root.mainloop()
````

## Solution 1

````python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.status = "Available"
    
    def show_info(self):
        print(f'{self.title} by {self.author}: {self.status}')

    def borrow(self):
        if self.status == "Available":
            self.status = "Borrowed"
            print(f"You borrowed '{self.title}'")
        else:
            print(f"'{self.title}' is already borrowed.")

    def return_book(self):
        if self.status == "Borrowed":
            self.status = "Available"
            print(f"Returned '{self.title}'")
        else:
            print(f"'{self.title}' was not borrowed.")
````

````python
class User:
    def __init__(self, username, email, password):
        self.username = username
        self.email = email
        self.__password = password
        self.is_logged_in = False

    def login(self, password):
        if password == self.__password:
            self.is_logged_in = True
            print(f"{self.username} logged in successfully.")
        else:
            print("Incorrect password. Login failed.")

    def logout(self):
        if self.is_logged_in:
            self.is_logged_in = False
            print(f"{self.username} has been logged out.")
        else:
            print("You are not logged in.")

    def reset_password(self, old_password, new_password):
        if self.is_logged_in:
            if old_password == self.__password:
                self.__password = new_password
                print("Password updated successfully.")
            else:
                print("Old password is incorrect. Password not changed.")
        else:
            print("Access denied. Please log in to reset password.")

    def get_info(self):
        if self.is_logged_in:
            print("User Information:")
            print(f"Username: {self.username}")
            print(f"Email: {self.email}")
        else:
            print("Access denied. Please log in to view information.")
````
